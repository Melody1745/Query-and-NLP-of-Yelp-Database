```python
# insert user dataset
def userdata_insert(db):

    with open('./yelp_academic_dataset_user.json', 'r', encoding='utf-8') as f:
        i = 0
        while True:
            i += 1
            try:
                # insert data line by line
                lines = f.readline()
                user_text = json.loads(lines)
                result = []
                result.append((user_text['user_id'], user_text['name'], user_text['review_count'], user_text['yelping_since'], 
                               user_text['useful'], user_text['funny'], user_text['cool'], user_text['elite'], user_text['friends']))
                insert_user = "insert into user(user_id, name, review_count, yelping_since, useful, funny, cool, elite, friends) values (%s, %s, %s, %s, %s, %s, %s, %s, %s)"
                cursor = db.cursor()
                cursor.executemany(insert_user, result)
                db.commit()
            except Exception as e:
                db.rollback()
                print(str(e))
                break

#initialize
if __name__ == "__main__":
    db = pymysql.connect(host="yelp.ciyx8eht0ylz.us-east-1.rds.amazonaws.com", user="melody", password="19931206Wr.", database="yelp", charset='utf8', port=3306)
    cursor = db.cursor()
    userdata_insert(db)
    cursor.close()
```

```python
# insert business dataset
db = pymysql.connect(host="yelp.ciyx8eht0ylz.us-east-1.rds.amazonaws.com", port=3306, user="melody", password="19931206Wr.", database="yelp", charset='utf8')
cursor = db.cursor()
with open('./yelp_academic_dataset_business.json', 'r', encoding='utf-8') as f:
    i=0
    for line in f.readlines():
        i +=1
        dic = json.loads(line)
        keys = ', '.join(dic.keys())
        valuesList = [dici for dici in dic.values()]
        
        #convert the nested json data to str
        for index1, value1 in enumerate(valuesList):
            if isinstance(value1, dict):
                valuesList[index1] = str(value1)
            elif isinstance(value1, list):
                valuesList[index1] = str(value1)
        
        # import data as tuple
        valuesTuple = tuple(valuesList)
        values = ', '.join(['%s']*len(dic)) 
        table="business"
        sql = 'INSERT INTO {table}({keys}) VALUES ({values})'.format(table=table, keys=keys, values=values)  
        cursor.execute(sql,valuesTuple)
        db.commit()
db.close()
```
